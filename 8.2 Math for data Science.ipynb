 1. Solving a System of Linear Equations
Use Python to solve the following system of equations. Hint: You can write this system as a matrix
equation in in the form Ax = b, where the matrix A is invertible

27x1 − 10x2 + 4x3 − 29x4 = 1

−16x1 + 5x2 − 2x3 + 18x4 = −1

−17x1 + 4x2 − 2x3 + 20x4 = 0

−7x1 + 2x2 − x3 + 8x4 = 1

```python
import numpy as np

A = np.matrix ([[27, -10, 4, -29],[-16, 5, -2, 18],[-17, 4, -2, 20],[-7, 2, -1, 8]])
A
```

```python
b = np.matrix([[1],[-1],[0],[1]])
b
```

```python
x = np.linalg.solve(A,b)
x
```

```python
# converting float to int
(A. dot(x)).astype(np.int32)  == np.matrix([[1],[-1],[0],[1]])
```
